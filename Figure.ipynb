```python
import xgi
import matplotlib.pyplot as plt

input_file = "E:\\CHARIS\\data.txt"

hyperedges = {}

with open(input_file, 'r', encoding='utf-8') as file:
    for line_number, line in enumerate(file, 1):
        parts = line.strip().split(": ")
        if len(parts) == 2:
            edge_id, nodes = parts
            edge_key = edge_id.strip()  
            nodes_set = set(nodes.split(",")) 
            hyperedges[edge_key] = nodes_set
        else:
            print(f"Line {line_number} does not contain ': ': {line.strip()}")

H = xgi.Hypergraph(hyperedges)
pos = xgi.barycenter_spring_layout(H)
ax, collections = xgi.draw(
    H,
    pos=pos,
    node_fc=[H.degree(node) for node in H.nodes], 
    edge_fc=[H.size(edge) for edge in H.edges],    
    node_labels=False,                             
    node_size=25                                
  
)

for node, (x, y) in pos.items():
    ax.text(x, y, node, fontsize=15, ha='center', va='center')


(node_collection, dyad_collection, edge_collection) = collections

plt.tight_layout()

plt.show()
```
